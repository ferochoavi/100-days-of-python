<h1>Funciones con inputs</h1>

```python
def greet_with_name(name):
    print(f"Hola {name}!")

greet_with_name("Fernando")
```

<h1>Funciones con mas de un input</h1>

```python
def greet_with(name, location):
    print(f"Hello {name}!")
    print(f"How is it in {location}?")

greet_with("Fernando", "Apodaca")
```

<h4>Lo malo de los argumentos posicionales, es que si se introducen los argumentos en desorden, el codigo hará un sin sentido</h4>
<h4>Se puede corregir usando "keyword arguments" dándole un valor a la variable dentro de los parémetos</h4>

```python
greet_with(location = "Apodaca", name = "Fernando")
```
